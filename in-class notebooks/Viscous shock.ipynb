In this notebook, we study traveling wave/front solutions of the (viscous) Burgers equation

$$
    u_t + u u_x = \epsilon u_{xx}.
$$

If we substitute the traveling wave ansatz $u(x,t) = U(x-ct) = U(\xi)$ (where $c$ is the speed of the traveling wave), we get a 2nd-order ODE that can be written as the pair of 1st-order ODEs:

\begin{align}
    U'(\xi) & = V \\
    V'(\xi) & = \frac{1}{\epsilon}(U-c)V
\end{align}

The equilibrium points of this system consist of the line $V=0$.  The eigenvalues of the Jacobian along that line are
$$
\lambda_1 = 0 \ \ \ \ \ \lambda_2 = \frac{U-c}{\epsilon}.
$$

Thus if $U>c$ the equilibrium is unstable, while if $U<c$ it is stable.  As a result, we can have heteroclinic connections, which are paths that leave an unstable equilibrium point and eventually arrive at a stable equilibrium point.  These correspond to traveling front solutions, with the equilibrium point values of $U$ corresponding to the asymptotic solution states to the left and right of the front.  We see that the speed $c$ must lie in between these asymptotic values; in fact, due to symmetry (and as predicted by the Rankine-Hugoniot condition), $c$ is exactly the average of the left and right states.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate
from ipywidgets import interact, FloatSlider
import matplotlib
font = {'size'   : 15}
matplotlib.rc('font', **font)
```

```python
fig, axes = plt.subplots(1,2,figsize=(12,6))

c = 1.0
epsilon = 0.2

u = np.linspace(-1, 2., 50)
v = np.linspace(-1, 1., 50)
 
V, U = np.meshgrid(v, u)
du = V
dv = V*(U-c)/epsilon

axes[0].streamplot(V,U,dv,du,broken_streamlines=False,density=0.7)
axes[0].set_ylim(-1,2)
axes[0].set_xlabel('v'); axes[0].set_ylabel('u');
axes[0].plot([0,0],[-1,2],'-k',lw=3);

def rhs(t,w):
    u, v = w
    return np.array([v,v*(u-c)/epsilon])

w0 = np.array([1.4,-0.01])
t_eval = np.linspace(0,10,1000)
forwardsoln = scipy.integrate.solve_ivp(rhs,[0,10],w0,t_eval=t_eval)
backwardsoln = scipy.integrate.solve_ivp(rhs,[10,0],w0,t_eval=t_eval[::-1])
u = forwardsoln.y[0,:]
x = forwardsoln.t
axes[1].plot(x,u,'-r',lw=3)
axes[0].plot(np.diff(u)/np.diff(x),u[1:],'-r',lw=3)
u = backwardsoln.y[0,:][::-1]
x = -backwardsoln.t
axes[1].plot(x,u,'-r',lw=3)
axes[1].set_xlim(-10,10)
```

```python
fig, axes = plt.subplots(1,2,figsize=(12,6))
plt.close()

def viscous_shock(c=1.0,epsilon=0.2,u0=1.4):

    axes[0].cla()
    axes[1].cla()
    u = np.linspace(-1, 2., 50)
    v = np.linspace(-1, 1., 50)

    V, U = np.meshgrid(v, u)
    du = V
    dv = V*(U-c)/epsilon

    stream = axes[0].streamplot(V,U,dv,du,broken_streamlines=False,density=0.7)
    axes[0].set_ylim(-1,2)
    axes[0].set_xlabel('v'); axes[0].set_ylabel('u');
    axes[0].plot([0,0],[-1,2],'-k',lw=3);

    def rhs(t,w):
        u, v = w
        return np.array([v,v*(u-c)/epsilon])

    w0 = np.array([u0,-0.01])
    t_eval = np.linspace(0,10,1000)
    forwardsoln = scipy.integrate.solve_ivp(rhs,[0,10],w0,t_eval=t_eval)
    backwardsoln = scipy.integrate.solve_ivp(rhs,[10,0],w0,t_eval=t_eval[::-1])
    u = forwardsoln.y[0,:]
    x = forwardsoln.t
    axes[1].plot(x,u,'-r',lw=3)
    axes[0].plot(np.diff(u)/np.diff(x),u[1:],'-r',lw=3)
    u = backwardsoln.y[0,:][::-1]
    x = -backwardsoln.t
    axes[1].plot(x,u,'-r',lw=3)
    axes[1].set_xlim(-10,10)
    fig.canvas.draw_idle()
    plt.close()
    return fig
    
interact(viscous_shock,epsilon=FloatSlider(min=0.01,max=0.5,step=0.01,value=0.2));
```
